```python
moviesPerYear = data.groupby('Year').size()
plt.figure(figsize=(10,6))
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.plot(moviesPerYear.index, moviesPerYear.values, marker='o')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Runtime (Minutes)'], bins=100, color='skyblue', edgecolor='orange')
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime (Minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Rating'], bins=100, color='purple', edgecolor='black')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
correlation = data['Rating'].corr(data['Runtime (Minutes)'])
print("Correlation coefficient between movie ratings and runtimes:", correlation)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Rating'], data['Revenue (Millions)'], alpha=1)
plt.title('Relationship between Movie Ratings and Box Office Earnings')
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Revenue (Millions)'], bins=100, edgecolor='black')
plt.title('Distribution of Movie Budgets')
plt.xlabel('Revenue (Millions)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Rating'], bins=80, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Votes'], data['Rating'], alpha=0.5)
plt.title('Relationship between Number of Votes and Movie Ratings')
plt.xlabel('Number of Votes')
plt.ylabel('Movie Rating')
plt.show()
```
